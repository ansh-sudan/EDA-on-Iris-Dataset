```python
#importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Finding the outliers by using boxplot
import matplotlib.pyplot as plt

# Create a figure and axis
fig, ax = plt.subplots(figsize=(5, 4))

# Create the boxplot
ax.boxplot(df['sepal_length'], vert=False)

# Set labels and title
ax.set_xlabel('X-axis Label')
ax.set_title('Boxplot for sepal_length ')

# Show the plot
plt.show()
```

from above we can say that their is no outliers in sepal_length

```python
# scatter Plot 
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(df['sepal_length'],df['sepal_width'], color='blue', marker='^',alpha=0.7,edgecolors='red')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Scatter Plot between sepal_length and sepal_width ')
plt.grid(True)
plt.show()
```

from the above we can conclude that sepal length and width are more scatter 

```python
## scatter Plot
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(df['petal_length'],df['petal_width'], color='blue', marker='^',alpha=0.7,edgecolors='red')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Scatter Plot between petal_length and petal_width ')
plt.grid(True)
plt.show()
```

from the above we can conclude that petal length and width are forming the linear relation with each other

```python
import matplotlib.cm as cm
grouped = df.groupby('species')['sepal_width'].sum()
color_map = cm.get_cmap('tab20c', len(grouped))
grouped.plot(kind='bar',color=color_map(range(len(grouped))))
plt.title("Bar Chart showing sepal_width of different species")
plt.xlabel("Species")
plt.ylabel("sepal_width")
plt.show()
```

iris_ setosa have longest sepal width

```python
import seaborn as sns
import numpy as np
corr = df.corr()
mask = np.array(corr)
mask[np.tril_indices_from(mask)]= False
plt.figure(figsize=(5,5))
sns.heatmap(corr,linewidth=0.5,mask=mask,annot=True,square=True)
```
